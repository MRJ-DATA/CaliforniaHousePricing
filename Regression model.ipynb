```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

###Lets Load the California House Pricing Dataset

```python
from sklearn.datasets import fetch_california_housing
```

```python
california=fetch_california_housing()
```

```python
california.keys()
```

```python
## Lets check the DESC of the dataste
print(california.DESCR)
```

```python
print(california.data)
```

```python
print(california.target)
```

```python
print(california.feature_names)
```

```python
 ##Preparing The Dataset
```

```python
dataset = pd.DataFrame(california.data, columns=california.feature_names)
```

```python
dataset.head()
```

```python
dataset['Price']=california.target
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
## Summarizing The Stats of the data
dataset.describe()
```

```python
## Summarizing The Stats of the data
dataset.describe()
```

```python
### EDA Exploratory Data Analysis 
## Correlation

dataset.corr()
```

```python
plt.scatter(dataset['AveRooms'],dataset['Price'])
```

```python
 import seaborn as sns
 sns.regplot(x="AveRooms",y="Price",data=dataset)
```

```python

 import seaborn as sns
 sns.regplot(x="Population",y="Price",data=dataset)
```

```python
 ## independent and dependent features

X=dataset.iloc[:,:-1]  #skip up all the independent features
Y=dataset.iloc[:,-1]
```

```python
X.head()
```

```python
Y
```

```python
## Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.3,random_state = 42)
```

```python
X_train
```

```python
X_test
```

```python
## Standardize the dataset   internally we use gradient desent out main aim is to come to the global minima and for that we need to make sure that all our independent unit scale should be in the same scale because of this the conversions will happen quickly
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
import pickle
pickle.dump(scaler,open('scaling.pkl','wb'))
```

```python
X_train
```

```python
X_test
```

```python
#Model Traning
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
```

```python
regression.fit(X_train,y_train)
```

```python
## print the coefficients and the intercept
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
## on which parameters the model has been trained
regression.get_params()
```

```python
###Prediction with Test Data
reg_pred=regression.predict(X_test)
```

```python
##Plot a scatter plot for the prediction
plt.scatter(y_test,reg_pred)
```

```python
##Residuals which are basically errors
residuals= y_test-reg_pred
```

```python
residuals
```

```python
## Plot this residuals

sns.displot(residuals, kind="kde")
```

```python
##Scatter plot wrt prediction and residuals
plt.scatter(reg_pred,residuals)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_absolute_error(y_test,reg_pred))

print(mean_squared_error(y_test,reg_pred))
```

## R square and adjusted R square

```python
R^2 = SSR/SST
 
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,reg_pred)
print(score)
```

```python
1 - (1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```
